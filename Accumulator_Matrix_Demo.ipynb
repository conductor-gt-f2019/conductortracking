# Accumulator Matrix Demo

```python
import numpy as np
import accumulator as acc
```

## Load Example File

```python
example_file = '2-4_001.csv'
```

## Load Coordinates from File

```python
coords = acc.load_coords_from_file(example_file, hand=1)
print('{:,} points loaded from "{}".'.format(coords.shape[0], example_file))
```

## Create Accumulator Matrix

```python
block_size = 50
acc_matrix = acc.create_accumulator_matrix(coords, block_size=block_size)
```

## Show Image

```python
acc.show_acc_matrix(acc_matrix)
```

## Create Multiple Accumulator Matrices via Blocks

```python
# create block windows
blocks = acc.create_block_windows(coords.shape[0], block_size)
# create multiple accumuator matrices via block windows
acc_matrices = acc.create_multiple_acc_matrices(coords, blocks)
```

## Display Multiple Accumulator Matrices

```python
for a_m in acc_matrices:
    acc.show_acc_matrix(a_m)
```
